## Data Processing

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler, RobustScaler
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, LabelEncoder
```

```python
df_prep=df.copy()
df_prep["sepal length"]=StandardScaler.fit_transform(X=df["sepal length"])
#df_prep["sepal width"]=StandardScaler.fit_transform(df_prep["sepal width"])
```

```python
from sklearn.compose import ColumnTransformer
clt=ColumnTransformer(
    transformers=[('std', StandardScaler(), [0,1]),
                ('minmax', MinMaxScaler(), [2,3]),
                ('cat', LabelEncoder(), [4])],
    remainder='drop'
)

clt.fit_transform(df_prep)
```
