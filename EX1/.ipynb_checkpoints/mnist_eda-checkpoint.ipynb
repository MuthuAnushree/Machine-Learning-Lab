```python
from keras.datasets import mnist
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

```python
print('X_train: ' + str(train_X.shape))
print('Y_train: ' + str(train_y.shape))
print('X_test:  '  + str(test_X.shape))
print('Y_test:  '  + str(test_y.shape))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(train_X[i], cmap="gray")
    plt.title(f"Label: {train_y[i]}")

    plt.axis("off")
plt.suptitle("Images from Dataset")
plt.show()
```

```python
unique, counts = np.unique(train_y, return_counts=True)

plt.bar(unique, counts)
plt.xlabel("Digit Class")
plt.ylabel("Count")
plt.title("Class Distribution in Training Set")
plt.show()
```

```python
plt.hist(train_X.flatten(), bins=50)
plt.title("Pixel Intensity Distribution")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Mean pixel value:", np.mean(train_X))
print("Std pixel value:", np.std(train_X))
```

```python
plt.figure(figsize=(10,4))
for digit in range(10):
    avg_img = np.mean(train_X[train_y == digit], axis=0)
    plt.subplot(2,5,digit+1)
    plt.imshow(avg_img, cmap='gray')
    plt.title(str(digit))
    plt.axis('off')
plt.suptitle("Average Image per Digit")
plt.show()
```

```python
from sklearn.decomposition import PCA

x_flat = train_X.reshape(len(train_X), -1)
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_flat[:5000])

plt.scatter(x_pca[:,0], x_pca[:,1], c=train_y[:5000], cmap='tab10', s=5)
plt.colorbar()
plt.title("PCA Projection of MNIST")
plt.show()
```
