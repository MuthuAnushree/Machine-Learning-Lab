```python
#### Change percentile or try and cap the values.....

cols = df.columns
for col in cols:
    q1 = df[col].quantile(0.01)
    q3 = df[col].quantile(0.99)
    df[col] = np.clip(df[col],q1,q3)
    
df.describe().T
```
