```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
from sklearn.model_selection import train_test_split
Y=df['class']
X=df.drop('class', axis=1)
X_train, X_test, y_train, y_test=train_test_split(X, Y, random_state=42, test_size=0.2)
print(f"Test Size: {len(X_test)}")
```

```python
y_test.shape
```

```python
gaussnb=GaussianNB()
gaussnb.fit(X=X_train, y=y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score

#write function to store all error metrics in a df

y_pred=gaussnb.predict(X_test)
accuracy=accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
plt.imshow(cm, cmap="Blues")    
plt.title("Gaussian NB - Confusion Matrix")
plt.xlabel("Predicted Label", fontsize=12)
plt.ylabel("Actual Label", fontsize=12)
plt.savefig('Screenshots/NB/Gaussian/cm.png')
```

```python
def store_metrics(metrics_df, model_name, model):
    start=time.perf_counter()
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    end=time.perf_counter()
    duration=end-start
    t=y_test
    p=y_pred
    length=len(metrics_df)
    tn, fp, fn, tp = confusion_matrix(t,p).ravel()
    specificity=tn/(tn+fp)
    metrics_df.loc[length+1]=([model_name, accuracy_score(t, p), precision_score(t,p), recall_score(t,p), f1_score(t,p), specificity, duration])
```

```python
cols=['model', 'accuracy', 'precision', 'recall', 'f1_score', 'specificity', 'time_taken']
nb_metrics = pd.DataFrame(columns=cols)
```

```python
start=time.perf_counter()
gaussnb=GaussianNB()
gaussnb.fit(X=X_train, y=y_train)
y_pred=gaussnb.predict(X_test)
end=time.perf_counter()
t=y_test
p=y_pred
duration=end-start
print("GaussNB Accuracy: ", accuracy_score(y_test, y_pred))
print("GaussNB Precision: ", precision_score(y_test, y_pred))
print("Time taken; ", duration , " s")
store_metrics(nb_metrics, 'GaussianNB', GaussianNB())
```

```python
from sklearn.metrics import roc_curve, auc

y_prob=gaussnb.predict_proba(X_test)[:,1]
model=gaussnb
plt.figure(figsize=(7, 5))

fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f'{model} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'r--', label='Random Guess')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.savefig('Screenshots/NB/Gaussian/roc.png')
plt.show()
```

```python
def store_metrics2(metrics_df, model_name, model):
    #training
    t_start=time.perf_counter()
    model.fit(X_train, y_train)
    t_end=time.perf_counter()
    t_duration=t_end-t_start
    #predicting
    p_start=time.perf_counter()
    y_pred=model.predict(X_test)
    p_end=time.perf_counter()
    p_duration=p_end-p_start
    t=y_test
    p=y_pred
    length=len(metrics_df)
    tn, fp, fn, tp = confusion_matrix(t,p).ravel()
    specificity=tn/(tn+fp)
    metrics_df.loc[length+1]=([model_name, accuracy_score(t, p), precision_score(t,p), recall_score(t,p), f1_score(t,p), specificity, t_duration, p_duration])
```

```python
cols=['model', 'accuracy', 'precision', 'recall', 'f1_score', 'specificity', 'training_time', 'prediction_time']
knn_metrics = pd.DataFrame(columns=cols)
```

```python
store_metrics2(knn_metrics, "KNN", KNeighborsClassifier())
```

```python
acc_scores=[]
for i in range(1,31):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    acc=accuracy_score(y_test, y_pred)
    acc_scores.append(acc)

sns.lineplot(x=range(1,31), y=acc_scores)
plt.title("KNN Accuracy Curve")
plt.xlabel("k values")
plt.ylabel("Accuracy")
plt.savefig('Screenshots/KNN/KNN/acc.png')
plt.show()
```
